```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
from  sklearn.cluster import DBSCAN
```

```python
two_blob=pd.read_csv('C:/SINGH/Data/cluster_two_blobs.csv')
```

```python
two_blobs_outliers=pd.read_csv('C:/SINGH/Data/cluster_two_blobs_outliers.csv')
```

```python
two_blob
```

```python
sns.scatterplot(data=two_blob,x='X1',y='X2')
```

```python
sns.scatterplot(data=two_blobs_outliers,x='X1',y='X2')
```

```python
def display_categories(model,data):
    labels=model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels,palette='Set1')
```

```python
model=DBSCAN()
```

```python
display_categories(model,two_blob)
```

```python
display_categories(model,two_blobs_outliers)
```

```python
dbscan= DBSCAN(eps=1)
```

```python
display_categories(dbscan,two_blobs_outliers)
```
